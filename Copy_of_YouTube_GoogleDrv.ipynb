<h4><b>YouTube to Google Drive Downloader (Colab)</h4></b><br>
This Jupyter Notebook will guide you through the process of downloading a YouTube video and saving it directly to your Google Drive.

```python
# 2. Mount Google Drive
from google.colab import drive

# This will prompt for authorization.
drive.mount('/content/drive')
```

```python

import yt_dlp
import os

# Define where to save the video in your Google Drive
# This creates a folder named "YouTube_Downloads" in your main Drive directory
save_path = '/content/drive/MyDrive/YouTube_Downloads'
os.makedirs(save_path, exist_ok=True)

# Prompt for the YouTube link
video_url = input("Enter the YouTube link: ")

# Configure the download options
ydl_opts = {
    # Save format: Title of the video followed by the extension
    'outtmpl': f'{save_path}/%(title)s.%(ext)s',

    # Download the best single file quality (usually mp4)
    'format': 'best',

    # Ignore errors to prevent the script from crashing on minor issues
    'ignoreerrors': True,
    'no_warnings': True
}

# Execute the download
print("\nStarting download... This might take a moment depending on the video length.")

try:
    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        ydl.download([video_url])
    print(f"\n✅ Success! The video has been saved to your Google Drive in the 'YouTube_Downloads' folder.")
except Exception as e:
    print(f"\n❌ An error occurred during the download: {e}")
```

```python
from pytube import YouTube
import os

# @param {type:"string"} youtube_url
youtube_url = "https://youtu.be/Yb8pGWKzpkU?si=21WvwJ6JPJt7A8Xf" # Replace with your YouTube video URL

# Download path in Google Drive
download_path = '/content/drive/My Drive/'

try:
    yt = YouTube(youtube_url)
    # Select the best quality stream, you can modify this as needed
    stream = yt.streams.get_highest_resolution()
    print(f"Downloading: {yt.title}...")
    stream.download(output_path=download_path)
    print("Download complete!")
except Exception as e:
    print(f"An error occurred: {e}")
```

This notebook will download the video directly to your specified Google Drive path after you run the cells sequentially and authorize access. Remember to replace "YOUR_YOUTUBE_URL" with the actual YouTube link.
