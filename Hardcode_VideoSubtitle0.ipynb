```python
!apt-get install imagemagick
# Fix ImageMagick security policy to allow text-to-image conversion
!sed -i 's/domain="coder" rights="none" pattern="PDF"/domain="coder" rights="read|write" pattern="PDF"/' /etc/ImageMagick-6/policy.xml
!sed -i 's/domain="coder" rights="none" pattern="label"/domain="coder" rights="read|write" pattern="label"/' /etc/ImageMagick-6/policy.xml
!sed -i 's/domain="path" rights="none" pattern="@\*"/domain="path" rights="read|write" pattern="@*"/' /etc/ImageMagick-6/policy.xml

import os
import re
from moviepy.config import change_settings

# Configure MoviePy to use ImageMagick
change_settings({"IMAGEMAGICK_BINARY": "/usr/bin/convert"})

from moviepy.editor import VideoFileClip, TextClip, CompositeVideoClip
from moviepy.video.tools.subtitles import SubtitlesClip

# File paths
video_path = "/my_video.mp4"
original_subtitle_path = "/content/ImpactChallenges.srt.txt"
clean_subtitle_path = "/content/cleaned_subtitles.srt"

# Validation check
if not os.path.exists(video_path):
    print(f"Error: Video file '{video_path}' not found.")
elif not os.path.exists(original_subtitle_path):
    print(f"Error: Subtitle file '{original_subtitle_path}' not found.")
else:
    try:
        # Read and Clean Subtitles
        with open(original_subtitle_path, 'r', encoding='utf-8-sig') as f:
            content = f.read()

        # Fix formatting: Ensure single empty line between blocks and remove excess whitespace
        content = re.sub(r'\n\s*\n+', '\n\n', content.strip())

        with open(clean_subtitle_path, 'w', encoding='utf-8') as f:
            f.write(content)

        print("Subtitles cleaned and saved to:", clean_subtitle_path)

        # Define the subtitle generator
        generator = lambda txt: TextClip(txt, font='Arial', fontsize=24, color='white', method='caption', size=(800, None))

        # Load video and subtitles
        video = VideoFileClip(video_path)
        subtitles = SubtitlesClip(clean_subtitle_path, generator)

        # Overlay subtitles on video (centered at the bottom)
        result = CompositeVideoClip([video, subtitles.set_pos(('center', 'bottom'))])

        # Render the final video
        output_filename = "burned_video.mp4"
        result.write_videofile(output_filename, fps=video.fps, codec="libx264")
        print(f"Success! Video saved as {output_filename}")

    except Exception as e:
        print(f"An error occurred: {e}")
        import traceback
        traceback.print_exc()
```
