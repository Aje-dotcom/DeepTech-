### 1\. Key advantages of using cloud computing for machine learning projects

According to the document.(Reading 1), the key advantages include:

- **Scalability**: You can scale resources up or down easily.
- **Cost-effectiveness**: You pay only for what you use.
- **Flexibility**: You can adapt to changing needs quickly.

### 2\. Public vs. Private Cloud and handling sensitive data

- A public cloud provides shared resources that are available to everyone.
- A private cloud provides dedicated resources exclusively for one organization.
- A **private cloud** is more suitable for an organization handling sensitive data because it is directly tied to providing "Organization Security".

### 3\. Main function of Azure Machine Learning in simplifying workflows

- Azure Machine Learning simplifies workflows by serving as a cloud-based platform that streamlines the creation, training, deployment, and management of machine learning models.
- It achieves this by providing a unified workspace to handle end-to-end workflows, from data preparation to model deployment and monitoring.

### 4\. Key benefits of using Azure Machine Learning for training and deploying models

The presentation lists the following specific benefits for Azure ML:

- Scalability.
- Automation.
- Real-time Predictions.
- Integration with the Azure Ecosystem.
- Cost Efficiency.
- Security and Compliance.

### 5\. Resources needed to set up an Azure Machine Learning workspace

To set up an Azure ML workspace, you need to prepare specific dependency resources that integrate into the workspace to ensure security, scalability, and resource management. These include:

- **Azure Storage**: To store data used in workflows, such as datasets, model artifacts, and logs.
- **Azure Container Registry**: To store Docker images used for training and deploying ML models.
- **Azure Key Vault**: To store and manage sensitive information, such as secrets, keys, and certificates.
- **Azure App Insights**: To collect telemetry data (like request rates and response times) to ensure reliable operation and performance.
