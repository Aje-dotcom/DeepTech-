```python
import os
import json
from google.colab import files
from markitdown import MarkItDown
from IPython.display import display, Markdown

# 1. Upload files from local machine
print("Please upload your PDF, DOCX, or other documents:")
uploaded = files.upload()

md_converter = MarkItDown()

for filename in uploaded.keys():
    print(f'\n--- Processing: {filename} ---')

    try:
        # 2. Convert to Markdown
        result = md_converter.convert(filename)
        md_content = result.text_content

        # DISPLAY CONTENT
        print("\nConverted Markdown Content:")
        display(Markdown(md_content))

        # 3. Save as .md
        base_name = os.path.splitext(filename)[0]
        md_filename = f"{base_name}.md"
        with open(md_filename, 'w', encoding='utf-8') as f:
            f.write(md_content)
        print(f"File saved: {md_filename}")

        # 4. Save as .ipynb (Notebook)
        notebook = {
            "cells": [
                {
                    "cell_type": "markdown",
                    "metadata": {},
                    "source": [f"# Converted from {filename}\n\n"] + [line + "\n" for line in md_content.splitlines()]
                }
            ],
            "metadata": {},
            "nbformat": 4,
            "nbformat_minor": 4
        }

        ipynb_filename = f"{base_name}.ipynb"
        with open(ipynb_filename, 'w', encoding='utf-8') as f:
            json.dump(notebook, f, indent=1)
        print(f"Notebook created: {ipynb_filename}")

    except Exception as e:
        print(f"Error processing {filename}: {e}")
```
