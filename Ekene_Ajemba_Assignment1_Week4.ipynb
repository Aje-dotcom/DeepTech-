### 2. Convert to 1NF (First Normal Form)
Requirement: Ensure atomicity (one value per cell) and remove duplicates.
In my data,<b><p>Step 1: Converting the dataset into 1NF</b><br>
To convert the dataset into<b> First Normal Form, </b></br>I ensured that:
* each column contains only <b>atomic values<br></b>
* there are no repeating groups<br>
* duplicate rows are removed<br><p>
The original dataset already had atomic fields such as emp_id, emp_name, department, salary, department_location, and manager_id, so the main adjustment was removing duplicate rows.
The resulting <b>1NF table is:</b>

```python
query_1nf = """
CREATE TABLE Employee_1NF AS
SELECT DISTINCT
    emp_id,
    emp_name,
    department,
    salary,
    department_location,
    manager_id
FROM EmployeeData;
"""
conn.execute("DROP TABLE IF EXISTS Employee_1NF")
conn.execute(query_1nf)

display(pd.read_sql("SELECT * FROM Employee_1NF", conn))
```

This table satisfies 1NF because every attribute contains only a single value and there are no duplicate records.

### Transforming the dataset into 2NF
To achieve Second Normal Form, I removed partial dependency by separating employee details from department-related details.

The reason for this is that department_location and manager_id depend on the department, not directly on the employee.

So the dataset was decomposed into the following tables:

```python
conn.execute("DROP TABLE IF EXISTS Employees_2NF")
conn.execute("DROP TABLE IF EXISTS Departments_2NF")

query_employees_2nf = """
CREATE TABLE Employees_2NF AS
SELECT DISTINCT
    emp_id,
    emp_name,
    department,
    salary
FROM Employee_1NF;
"""

query_departments_2nf = """
CREATE TABLE Departments_2NF AS
SELECT DISTINCT
    department,
    department_location,
    manager_id
FROM Employee_1NF;
"""

conn.execute(query_employees_2nf)
conn.execute(query_departments_2nf)

print("Employees_2NF")
display(pd.read_sql("SELECT * FROM Employees_2NF", conn))

print("Departments_2NF")
display(pd.read_sql("SELECT * FROM Departments_2NF", conn))
```

This step separates employee details from department details.

```python
conn.execute("DROP TABLE IF EXISTS Employees_3NF")
conn.execute("DROP TABLE IF EXISTS Departments_3NF")
conn.execute("DROP TABLE IF EXISTS Managers_3NF")

query_employees_3nf = """
CREATE TABLE Employees_3NF AS
SELECT DISTINCT
    emp_id,
    emp_name,
    department,
    salary
FROM Employee_1NF;
"""

query_departments_3nf = """
CREATE TABLE Departments_3NF AS
SELECT DISTINCT
    department,
    department_location,
    manager_id
FROM Employee_1NF;
"""

query_managers_3nf = """
CREATE TABLE Managers_3NF AS
SELECT DISTINCT
    manager_id
FROM Employee_1NF;
"""

conn.execute(query_employees_3nf)
conn.execute(query_departments_3nf)
conn.execute(query_managers_3nf)

print("Employees_3NF")
display(pd.read_sql("SELECT * FROM Employees_3NF", conn))

print("Departments_3NF")
display(pd.read_sql("SELECT * FROM Departments_3NF", conn))

print("Managers_3NF")
display(pd.read_sql("SELECT * FROM Managers_3NF", conn))
```

This *3NF* step removes transitive dependency by separating managers into their own table

✅Historical insights

Operational databases usually focus on current transactions, but a warehouse keeps historical data, which helps management study patterns over time such as monthly sales growth, customer behavior, and seasonal demand.

✅Improved data quality

Data warehousing involves cleaning and standardizing data, reducing inconsistencies and making reports more reliable.

✅Supports strategic decisions

### 2. SQL View for customers who have spent more than 300,000 Naira
Assuming there is a transaction table called `sales` with:

* customer_id
* amount

and a customer table called _customers_ with:

* customer_id
* name

the SQL View would be:

```python
sql_view_query = """
CREATE VIEW high_value_customers AS
SELECT
    c.customer_id,
    c.name,
    SUM(s.amount) AS total_spent
FROM customers c
JOIN sales s
    ON c.customer_id = s.customer_id
GROUP BY c.customer_id, c.name
HAVING SUM(s.amount) > 300000;
"""

conn.execute(sql_view_query)
```

This view returns only customers whose total purchases exceed 300,000 Naira.

### 3. Materialized View for total sales per month

Assuming:

##1. SQL script to create the tables with constraints

```python
import sqlite3

# Establish connection to an SQLite database file for Assignment 3
conn = sqlite3.connect('ordermanagement_assignment3.db')

# These commands are for other SQL dialects (like MySQL) and are not needed for SQLite.
# CREATE DATABASE OrderManagement;
# USE OrderManagement;

# Drop tables if they exist to allow re-running the cell
conn.execute("DROP TABLE IF EXISTS OrderDetails;")
conn.execute("DROP TABLE IF EXISTS Orders;")
conn.execute("DROP TABLE IF EXISTS Products;")
conn.execute("DROP TABLE IF EXISTS Customers;")

conn.execute("""
CREATE TABLE Customers (
    CustomerID INTEGER PRIMARY KEY AUTOINCREMENT,
    FirstName TEXT NOT NULL,
    LastName TEXT NOT NULL,
    Email TEXT UNIQUE NOT NULL,
    Phone TEXT
    -- SQLite does not natively support REGEXP in CHECK constraints without custom functions.
    -- CONSTRAINT chk_phone
    --     CHECK (Phone REGEXP '^[0-9]{10,15}$')
);
""")

conn.execute("""
CREATE TABLE Products (
    ProductID INTEGER PRIMARY KEY AUTOINCREMENT,
    ProductName TEXT NOT NULL,
    Price DECIMAL(10,2),
    StockQuantity INT,
    CONSTRAINT chk_price
        CHECK (Price > 0),
    CONSTRAINT chk_stock
        CHECK (StockQuantity >= 0)
);
""")

conn.execute("""
CREATE TABLE Orders (
    OrderID INTEGER PRIMARY KEY AUTOINCREMENT,
    CustomerID INT,
    OrderDate DATE NOT NULL,
    TotalAmount DECIMAL(10,2),
    CONSTRAINT chk_totalamount
        CHECK (TotalAmount > 0),
    CONSTRAINT fk_orders_customer
        FOREIGN KEY (CustomerID) REFERENCES Customers(CustomerID)
);
""")

conn.execute("""
CREATE TABLE OrderDetails (
    OrderDetailID INTEGER PRIMARY KEY AUTOINCREMENT,
    OrderID INT,
    ProductID INT,
    Quantity INT,
    Subtotal DECIMAL(10,2),
    CONSTRAINT chk_quantity
        CHECK (Quantity > 0),
    CONSTRAINT chk_subtotal
        CHECK (Subtotal > 0),
    CONSTRAINT fk_orderdetails_order
        FOREIGN KEY (OrderID) REFERENCES Orders(OrderID) ON DELETE CASCADE,
    CONSTRAINT fk_orderdetails_product
        FOREIGN KEY (ProductID) REFERENCES Products(ProductID)
);
""")

conn.commit()
print("Tables created successfully.")
```

This script creates all four tables from the Excel file and applies the constraints exactly as described:

* PRIMARY KEY for unique row identification
* AUTO_INCREMENT for automatic IDs
* NOT NULL to prevent missing required values
* UNIQUE on customer email
* CHECK constraints for phone, price, stock, quantity, subtotal, and total amount
* FOREIGN KEY constraints to enforce relationships between tables
* ON DELETE CASCADE on OrderDetails.OrderID so deleting an order also removes its details

2. SQL Transaction to Handle an Order Purchase

```python
import sqlite3

# Assuming 'conn' is already established (from previous cells)
# If not, you would need to establish it, e.g., conn = sqlite3.connect('ordermanagement_assignment3.db')

def process_order_transaction(customer_id, product_id, quantity):
    try:
        # Start a transaction (sqlite3 handles this implicitly with commit/rollback
        # but an explicit BEGIN can be used for clarity if needed)
        conn.execute("BEGIN TRANSACTION;")

        # Step 1: Check stock and get product price
        # Using a single SELECT statement to get both price and stock
        cursor = conn.execute(
            "SELECT Price, StockQuantity FROM Products WHERE ProductID = ?;",
            (product_id,)
        )
        result = cursor.fetchone()

        if result is None:
            print(f"Product with ID {product_id} not found.")
            conn.rollback()
            return

        product_price, available_stock = result

        # Step 2: Check if stock is sufficient
        if available_stock < quantity:
            print(f"Insufficient stock for Product ID {product_id}. Available: {available_stock}, Requested: {quantity}")
            conn.rollback()
            return

        # Step 3: Calculate total amount
        total_amount = product_price * quantity

        # Step 4: Insert into Orders table
        cursor = conn.execute(
            "INSERT INTO Orders (CustomerID, OrderDate, TotalAmount) VALUES (?, DATE('now'), ?);",
            (customer_id, total_amount)
        )
        order_id = cursor.lastrowid # Get the generated OrderID

        # Step 5: Insert into OrderDetails table
        conn.execute(
            "INSERT INTO OrderDetails (OrderID, ProductID, Quantity, Subtotal) VALUES (?, ?, ?, ?);",
            (order_id, product_id, quantity, total_amount)
        )

        # Step 6: Reduce stock in Products table
        conn.execute(
            "UPDATE Products SET StockQuantity = StockQuantity - ? WHERE ProductID = ?;",
            (quantity, product_id)
        )

        # If all steps succeed, commit the transaction
        conn.commit()
        print(f"Order {order_id} processed successfully for Customer {customer_id}, Product {product_id}, Quantity {quantity}.")

    except sqlite3.Error as e:
        print(f"An SQLite error occurred: {e}")
        conn.rollback()
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
        conn.rollback()

# --- Example Usage --- #
# First, let's ensure some dummy data exists for customers and products
# (You might want to run this in a separate setup cell or just ensure your previous setup created these)

# Clear existing data for a fresh run, if desired
# conn.execute("DELETE FROM OrderDetails;")
# conn.execute("DELETE FROM Orders;")
# conn.execute("DELETE FROM Products;")
# conn.execute("DELETE FROM Customers;")

# Add some sample customers
conn.execute("INSERT OR IGNORE INTO Customers (CustomerID, FirstName, LastName, Email, Phone) VALUES (?, ?, ?, ?, ?);", (1, 'Alice', 'Smith', 'alice@example.com', '1112223333'))
conn.execute("INSERT OR IGNORE INTO Customers (CustomerID, FirstName, LastName, Email, Phone) VALUES (?, ?, ?, ?, ?);", (2, 'Bob', 'Johnson', 'bob@example.com', '4445556666'))

# Add some sample products with initial stock
conn.execute("INSERT OR IGNORE INTO Products (ProductID, ProductName, Price, StockQuantity) VALUES (?, ?, ?, ?);", (101, 'Laptop', 1200.00, 10))
conn.execute("INSERT OR IGNORE INTO Products (ProductID, ProductName, Price, StockQuantity) VALUES (?, ?, ?, ?);", (102, 'Mouse', 25.00, 50))
conn.execute("INSERT OR IGNORE INTO Products (ProductID, ProductName, Price, StockQuantity) VALUES (?, ?, ?, ?);", (103, 'Keyboard', 75.00, 20))
conn.commit()

print("--- Initial Product Stock ---")
cursor = conn.execute("SELECT ProductID, ProductName, StockQuantity FROM Products;")
for row in cursor.fetchall():
    print(f"Product ID: {row[0]}, Name: {row[1]}, Stock: {row[2]}")

# Test Case 1: Successful purchase
print("\n--- Test Case 1: Successful purchase ---")
process_order_transaction(customer_id=1, product_id=101, quantity=2)

# Test Case 2: Insufficient stock
print("\n--- Test Case 2: Insufficient stock ---")
process_order_transaction(customer_id=2, product_id=101, quantity=15) # Should fail as only 8 remain (10 - 2)

# Test Case 3: Another successful purchase
print("\n--- Test Case 3: Another successful purchase ---")
process_order_transaction(customer_id=1, product_id=102, quantity=10)

print("\n--- Final Product Stock ---")
cursor = conn.execute("SELECT ProductID, ProductName, StockQuantity FROM Products;")
for row in cursor.fetchall():
    print(f"Product ID: {row[0]}, Name: {row[1]}, Stock: {row[2]}")

print("\n--- Orders ---")
cursor = conn.execute("SELECT OrderID, CustomerID, OrderDate, TotalAmount FROM Orders;")
for row in cursor.fetchall():
    print(f"Order ID: {row[0]}, Customer ID: {row[1]}, Date: {row[2]}, Total: {row[3]}")

print("\n--- Order Details ---")
cursor = conn.execute("SELECT OrderDetailID, OrderID, ProductID, Quantity, Subtotal FROM OrderDetails;")
for row in cursor.fetchall():
    print(f"Detail ID: {row[0]}, Order ID: {row[1]}, Product ID: {row[2]}, Quantity: {row[3]}, Subtotal: {row[4]}")
```

This transaction ensures the order process is atomic. That means all steps succeed together or none of them happen.

The transaction does the following:

checks the available stock,
creates a new order,
inserts the order details,
updates product stock,
commits the transaction if everything is valid.
If the requested quantity is greater than the available stock, the transaction performs a ROLLBACK, which cancels all changes and prevents an invalid sale.

3. Partition the Orders Table Using RANGE Partitioning

```python
conn.execute("DROP TABLE IF EXISTS Orders_Partitioned;")

conn.execute("""
CREATE TABLE Orders_Partitioned (
    OrderID INTEGER PRIMARY KEY AUTOINCREMENT,
    CustomerID INT,
    OrderDate DATE NOT NULL,
    TotalAmount DECIMAL(10,2),
    CONSTRAINT chk_totalamount_partitioned
        CHECK (TotalAmount > 0),
    CONSTRAINT fk_orders_partitioned_customer
        FOREIGN KEY (CustomerID) REFERENCES Customers(CustomerID)
);
""")
conn.commit()
print("Orders_Partitioned table created successfully (without partitioning as it's not supported in SQLite).")
```

This partitions the Orders table into two sections:

p_before_2023 stores orders before 2023
p_2023_and_after stores orders from 2023 onward
This helps when querying date-based records because the database can scan only the relevant partition instead of the whole table.

For example, if you search for orders placed in 2024, the database can focus only on the 2023_and_after partition

### 4. How Indexing Improved Query Performance
Indexes improve query performance by making data retrieval faster. Without an index, the database may need to scan every row in a table to find matching records. This is called a full table scan, and it becomes slow when the table is large.

For this schema, indexing is especially useful on foreign key and date columns.

Example indexes:

```python
conn.execute("CREATE INDEX idx_orders_customerid ON Orders(CustomerID);")
conn.execute("CREATE INDEX idx_orders_orderdate ON Orders(OrderDate);")
conn.execute("CREATE INDEX idx_orderdetails_orderid ON OrderDetails(OrderID);")
conn.execute("CREATE INDEX idx_orderdetails_productid ON OrderDetails(ProductID);")
conn.commit()
print("Indexes created successfully.")
```

These indexes improve performance in several ways.

An index on Orders(CustomerID) helps when searching for all orders made by a particular customer.

An index on Orders(OrderDate) helps for date filtering, sorting, and partition-related queries.

Indexes on OrderDetails(OrderID) and OrderDetails(ProductID) improve joins between Orders, OrderDetails, and Products.

For example, this query becomes faster with indexes:

```python
query = """
SELECT o.OrderID, o.OrderDate, od.ProductID, od.Quantity
FROM Orders o
JOIN OrderDetails od ON o.OrderID = od.OrderID
WHERE o.CustomerID = 1;
"""

df_result = pd.read_sql(query, conn)
display(df_result)
```

Instead of checking all rows in both tables, the database can use the index to quickly locate the needed records.

At the same time, indexing has a tradeoff. It improves SELECT performance, but it can slightly slow down INSERT, UPDATE, and DELETE operations because the indexes must also be updated whenever data changes
