
The SQL queries required to extract the requested information.

​1. Patients in "Fct Abuja" and "Plateau"
​This query uses the IN operator to filter the patient records based on their geographic location.

```python
SELECT * FROM Patients
WHERE State IN ('Fct Abuja', 'Plateau');
```

​2. Total Number of Male and Female Patients
​This query uses the COUNT aggregate function and GROUP BY to provide a demographic breakdown of your patient population

```python
SELECT Gender, COUNT(*) AS Total_Patients
FROM Patients
GROUP BY Gender;
```

3. Doctors with Confirmed Appointments
This query joins the Doctors and Appointments tables to identify specialists who have active, confirmed bookings. I have included DISTINCT to ensure each doctor is only listed once even if they have multiple confirmed appointments

```python
SELECT DISTINCT d.Name, d.Speciality, d.State
FROM Doctors d
JOIN Appointments a ON d.DoctorID = a.DoctorID
WHERE a.Status = 'Confirmed';
```
